```python
q1 = good.quantile(.25)
q3 = good.quantile(.75)
iqr = q3 - q1
iqr
```

```python
good_out = good[~((good < (q1 - 1.5 * iqr)) | (good > (q3 + 1.5*iqr))).any(axis=1)]
```
